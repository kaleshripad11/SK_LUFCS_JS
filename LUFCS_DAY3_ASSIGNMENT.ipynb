```python
#Program to take a number and define grades based on number
print("Please enter your marks out of 100 : ")
marks=int(input(""))
print("Your grades are as follows based on your marks :")
if((marks>0) and (marks<35)):
    print("Fail")
elif((marks>36) and (marks<50)):
    print("Grade D")
elif((marks>51) and (marks<65)):
    print("Grade C")
elif((marks>66) and (marks<80)):
    print("Grade B")
elif((marks>81) and (marks<100)):
    print("Grade A")

    
```

```python
#Guessing cricket score for 20-20
#random is an module that generates the random numbers within a range 
import random as randscore

#Generating random score and storing it to Score variable
score=randscore.randrange(1,250)
print("Todays score = ",score)

#Get users expectations for the score
guessed_score=int(input("Enter your expected score for todays 20-20 : "));
if((guessed_score<=1) | (guessed_score>=250)):
    print("Please reduce your score expectation for T-20")
else:
    #Next line calculate the difference between random score generated by system and score by user
    #abs() function is used to convert negative number in positive number
    approximate_score=abs(score-guessed_score)
    #print(approximate_score)
    if((approximate_score<=10) or (approximate_score>=1)):
        print("Close by, you are a true indian cricket fan!")
    else:
        print("You don't watch cricket that much!")
```
